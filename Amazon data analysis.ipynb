```python
#import helpful analytics libraries
import pandas as pd
import numpy as np  
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#Read the file and read the first and last five rows
df = pd.read_csv('/Users/zohrehsamieekadkani/Desktop/python-files/Abgabe/amazon.csv')    
print(df.head(5))
```

```python
#It cleans and converts specific columns in a Pandas DataFrame (df) from strings containing symbols (₹, %, ,, etc.) into proper numeric values.
for col in ['discounted_price',	'actual_price',	'discount_percentage','rating',	'rating_count']:
 df[col]=df[col].astype(str).str.replace('₹','').str.replace('%','').str.replace(',','').str.replace('nan','0')
 df[col] = pd.to_numeric(df[col].str.replace('[^0-9.]','',regex=True),errors='coerce').fillna(0)


print(df[['discounted_price','actual_price','discount_percentage']])
print(df[['discounted_price','actual_price','discount_percentage']].describe())
```

```python
sns.barplot(x=df['product_name'].head(10),y=df['rating_count'].head(10))
plt.xticks(rotation=90)
plt.title("Top 10 selling products as per rating count")
```

```python
split_columns = df['category'].str.split('|', expand=True)

# Rename the columns for clarity
split_columns.columns = [f'level_{i+1}' for i in range(split_columns.shape[1])]

# Combine the split columns with the original DataFrame (optional)
df = pd.concat([df, split_columns], axis=1)

df.head()
```

```python
df['level_1'].value_counts()
```

```python

df.groupby(by='level_1')['level_2'].value_counts().reset_index()
```

```python

df_cat_sell=df.groupby(by='level_1')['rating_count'].sum().sort_values(ascending=False).reset_index()
df_cat_sell
```

```python
sns.barplot(x=df_cat_sell['level_1'].head(10),y=df_cat_sell['rating_count'].head(10))
plt.xticks(rotation=90)
plt.title("Top 10 selling categories as per rating count")
```

```python
sns.barplot(x=df_cat_rating['level_1'].head(),y=df_cat_rating['rating'].head(5))
plt.xticks(rotation=90)
plt.title("Top 5 highest average rated categories")
```

```python
df_cat_dis_perc=df.groupby(by='level_1')['discount_percentage'].mean().sort_values(ascending=False).reset_index()
df_cat_dis_perc
```

```python
sns.barplot(x=df_cat_dis_perc['level_1'].head(),y=df_cat_dis_perc['discount_percentage'].head())
plt.xticks(rotation=90)
plt.title("Top 5 Categories which offer highest discount percentage")
```
